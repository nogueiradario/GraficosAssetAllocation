```python
import plotly.graph_objects as go
import pandas as pd
import colorsys

# =========================================
# 🔹 FUNÇÃO PARA CLAREAR COR
# =========================================

def clarear_cor(hex_color, fator=0.5):
    hex_color = hex_color.lstrip('#')
    r, g, b = [int(hex_color[i:i+2], 16)/255 for i in (0, 2, 4)]

    h, l, s = colorsys.rgb_to_hls(r, g, b)
    l = min(1, l + fator*(1-l))  # clareia

    r, g, b = colorsys.hls_to_rgb(h, l, s)
    return '#{:02x}{:02x}{:02x}'.format(int(r*255), int(g*255), int(b*255))

# =========================================
# 🔹 DADOS
# =========================================

dados = pd.DataFrame([
    ["Renda Variável","Ações Exterior","WRLD11",20.0],
    ["Renda Variável","Ações Brasil","BBOV11",7.5],
    ["Renda Variável","FIIs","ITIT11",2.5],
    ["Renda Fixa","Juros Real","IMAB11",15.0],
    ["Alternativos","Ouro","GOLD11",3.5],
], columns=["Macro","Classe","Ativo","Peso"])

# =========================================
# 🎨 CORES BASE
# =========================================

cores_macro = {
    "Renda Variável": "#7997A9",
    "Renda Fixa": "#97A979",
    "Alternativos": "#A97997"
}

# =========================================
# 🔹 CONSTRUÇÃO MANUAL DO SUNBURST
# =========================================

labels = []
parents = []
values = []
colors = []

# Macro
for macro in dados["Macro"].unique():
    total_macro = dados[dados["Macro"] == macro]["Peso"].sum()

    labels.append(macro)
    parents.append("")
    values.append(total_macro)
    colors.append(cores_macro[macro])

    # Classe
    for classe in dados[dados["Macro"] == macro]["Classe"].unique():
        df_classe = dados[(dados["Macro"] == macro) & (dados["Classe"] == classe)]
        total_classe = df_classe["Peso"].sum()

        labels.append(classe)
        parents.append(macro)
        values.append(total_classe)
        colors.append(clarear_cor(cores_macro[macro], 0.4))

        # Ativo
        for _, row in df_classe.iterrows():
            labels.append(row["Ativo"])
            parents.append(classe)
            values.append(row["Peso"])
            colors.append(clarear_cor(cores_macro[macro], 0.7))

# =========================================
# 📊 GRÁFICO
# =========================================

fig = go.Figure(go.Sunburst(
    labels=labels,
    parents=parents,
    values=values,
    marker=dict(colors=colors),

    texttemplate="%{label}<br>%{value:.1f}%",
    hovertemplate="<b>%{label}</b><br>%{value:.1f}%<extra></extra>"
))

fig.update_layout(margin=dict(t=50, l=0, r=0, b=0))

fig.show()
```
